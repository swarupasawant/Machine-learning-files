```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("F:\\Machine Learning\\data\\Social_Network_Ads.csv")
```

```python
dataset
```

```python
x= dataset.iloc[:,[2,3]].values
y= dataset.iloc[:,4].values
```

```python
plt.scatter(x[y==0,0],x[y==0,1],color="red",label="class-0")
plt.scatter(x[y==1,0],x[y==1,1],color="purple",label="class-1")
plt.legend()
plt.title("Logistic regression")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
#train test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25)
```

```python
#inverse of x_test
i= sc.inverse_transform(x_test)
i=pd.DataFrame(i)
```

```python
#Logistic fit the algorithm
from sklearn.linear_model import LogisticRegression
lgr_model=LogisticRegression()
lgr_model.fit(x_train,y_train)
```

```python
#prediction
y_pred=lgr_model.predict(x_test)
```

```python
y_pred
```

```python
#confusion matrix use for test accuracy in classification
from sklearn.metrics import confusion_matrix
mx=confusion_matrix(y_test,y_pred)
print(mx)
```

```python
from sklearn.metrics import accuracy_score
ac=accuracy_score(y_test,y_pred)
print(ac)
```

```python
from sklearn.metrics import classification_report
cr=classification_report(y_test,y_pred)
print(cr)
```

```python
#visualisation of train and test
from matplotlib.colors import ListedColormap
x_set,y_set=x_train,y_train

x1,x2=np.meshgrid(np.arange(start = x_set[:,0].min()-1,stop = x_set[:,0].max()+1, step =0.01),
                  np.arange(start = x_set[:,1].min()-1,stop = x_set[:,1].max()+1, step =0.01))
plt.contourf(x1 ,x2,lgr_model.predict (np.array([x1.ravel(),x2.ravel()]).T).reshape(x1.shape),
             alpha =0.75, cmap = ListedColormap(("pink","purple")))

plt.scatter(x_test[y_test==0,0],x_test[y_test==0,1],color="red",label="class-0")
plt.scatter(x_test[y_test==1,0],x_test[y_test==1,1],color="purple",label="class-1")


plt.xlim(x1.min(),x1.max())
plt.ylim(x1.min(),x1.max())
plt.title('Logistic Reg')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.show()

```
